```python
# import necessaty libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
```

```python
dataset = pd.read_csv('Nutrition__Physical_Activity__and_Obesity_-_Behavioral_Risk_Factor_Surveillance_System.csv')
dataset.sample(5)
```

```python
dataset['Question'].unique()
```

```python
dataset.info()
```

```python
dataset.describe(include = 'all')
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
# remove redundant columns

all_columns = dataset.columns.copy()
columns_to_remove = ['Low_Confidence_Limit','High_Confidence_Limit ','YearEnd','Topic','Class','Datasource','Data_Value_Unit','QuestionID','ClassID','TopicID','DataValueTypeID','Data_Value_Type','Data_Value_Footnote_Symbol','Data_Value_Footnote','StratificationCategoryId1','StratificationID1']

needed_column = []
for i in range(len(all_columns)):
    if all_columns[i] not in columns_to_remove:
        needed_column.append(all_columns[i])
    
needed_column
```

```python
dataset = dataset[needed_column]
```

```python
dataset.sample(5)
```

```python
# making use of the 2014 year data


dataset = dataset[dataset['YearStart']==2014]

# also seperating out data for just the obesity
dataset = dataset[dataset['Question'] == 'Percent of adults aged 18 years and older who have obesity']


```

```python
dataset['LocationDesc'].unique()
```

```python
# taking out dataset with just the states in the USA

dataset = dataset[~(dataset['LocationDesc']=='National')]
dataset = dataset[~(dataset['LocationDesc']=='Guam')]
dataset = dataset[~(dataset['LocationDesc']=='Puerto Rico')]
dataset['LocationDesc'].unique()
```

```python
# average percentage obesity 

average_percentage_obesity = round(dataset['Data_Value'].sum()/len(dataset['Data_Value']), 2)
average_percentage_obesity
```

```python
# average_percentage_obesity by state

sorted_dataset = dataset.sort_values('LocationDesc', ascending = False)
sorted_dataset=sorted_dataset[['LocationAbbr','LocationDesc','Data_Value','Gender']]

sorted_dataset.groupby('LocationDesc').mean()
```

```python
by_state = sorted_dataset.groupby('LocationDesc').mean()
by_state_sorted = by_state.sort_values('Data_Value', ascending = False)
by_state_sorted
```

```python
plt.rc('font', family = 'serif')
plt.rcParams.update({'font.size':14})

plt.figure(figsize = (10,16))

plt.subplot(2,1,1)
ax=sns.barplot(y=by_state_sorted.head(10).index, x=by_state_sorted.head(10).Data_Value, palette="Blues_d")
ax.set_ylabel('USA State')
ax.set_xlabel('Obesity Rate (%)')
ax.set_title('10 Most Obese States in 2014')

plt.plot([average_percentage_obesity,average_percentage_obesity],[-1,10], '--',color = 'r')

plt.subplot(2,1,2)
ax=sns.barplot(y=by_state_sorted.tail(10).index, x=by_state_sorted.tail(10).Data_Value, palette="Blues_d")
ax.set_ylabel('USA State')
ax.set_xlabel('Obesity Rate (%)')
ax.set_title('10 Least Obese States in 2014')

plt.plot([average_percentage_obesity,average_percentage_obesity],[-1,10], '--',color = 'r')
```

```python
# creating differendt dataframe by gender, education, income, race and age.


df_gender = dataset[(dataset['Stratification1']=='Male')|(dataset['Stratification1']=='Female')].reset_index(drop = True)
df_edu = dataset[dataset['StratificationCategory1']=='Education'].reset_index(drop = True)
df_income = dataset[dataset['StratificationCategory1']=='Income'].reset_index(drop = True)
df_race = dataset[dataset['StratificationCategory1']=='Race/Ethnicity'].reset_index(drop = True)
df_age = dataset[dataset['StratificationCategory1'] == 'Age (years)'].reset_index(drop = True)
```

```python
# To get the correlation of Gender with the percentage obesity

# select the needed columns
df_edu = df_edu[['LocationAbbr','LocationDesc','Data_Value','Education']]
```

```python
df_edu['Education'].unique()
```

```python
# to transform the education column to 

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

transformer = ColumnTransformer([('encoder', OneHotEncoder(), [3])], remainder = 'passthrough')
np.array(transformer.fit_transform(df_edu))
```

```python
# Transform the education column
#Create a list of all 4 educational levels
ledu=df_edu.Education.unique()

for i in ledu:
    df_edu[i]=df_edu['Education'].apply(lambda x: int(x==i))
    
df_edu
```

```python
df_edu_NHS = df_edu[df_edu["Less than high school"]==1].reset_index(drop = True)
sorted_df_edu_NHS = df_edu_NHS.groupby('LocationDesc')['Data_Value'].mean().sort_values(ascending = False)

sorted_df_edu_NHS
```

```python
NHS_average_percentage_obesity = df_edu_NHS['Data_Value'].sum() / len(df_edu_NHS['Data_Value']) 
round(NHS_average_percentage_obesity, 2)
```

```python
plt.rc('font', family = 'serif')
plt.rcParams.update({'font.size':14})

plt.figure(figsize = (10,16))

plt.subplot(2,1,1)
ax=sns.barplot(y=sorted_df_edu_NHS.head(10).index, x=sorted_df_edu_NHS.head(10), palette="Blues_d")
ax.set_ylabel('USA State')
ax.set_xlabel('Obesity Rate (%)')
ax.set_title('10 Most Obese States in 2014 Without High School')

plt.plot([NHS_average_percentage_obesity,NHS_average_percentage_obesity],[-1,10], '--',color = 'r')



plt.subplot(2,1,2)
ax=sns.barplot(y=sorted_df_edu_NHS.tail(10).index, x=sorted_df_edu_NHS.tail(10), palette="Blues_d")
ax.set_ylabel('USA State ')
ax.set_xlabel('Obesity Rate (%)')
ax.set_title('10 Least Obesed States in 2014 Without High School')

plt.plot([NHS_average_percentage_obesity,NHS_average_percentage_obesity],[-1,10], '--',color = 'r')
```

```python
# let's test the effect of education on obesity

# i will drop wyoming dataset because it is an outlier

df_edu = df_edu[~(df_edu['LocationDesc']=='Wyoming')]
                  
# also since we have three dummy column, we will be needing three coeficient so we will drop one
edu_cols = df_edu.columns[5:8]

X = df_edu[edu_cols].values
y = df_edu['Data_Value'].values

# let's split the dataset into train_set and test_set
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0, test_size = 0.2)



from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_train, y_train)

```

```python
regressor.coef_
```

```python


plt.barh([0,1,2],regressor.coef_)
plt.yticks(range(3),df_edu.columns[4:7], fontsize = 10)
plt.title('Regression Coefficients')

plt.show()
```

```python
print('R^2 on training...', regressor.score(X_train,y_train))
print('R^2 on test...', regressor.score(X_test,y_test))

print('Model Coefficients', regressor.coef_)
print('Model Intercept', regressor.intercept_)
```

```python
# Now let's look at the effecto of income on obesity

# selecting columns for the income data
df_income = df_income[['LocationAbbr','LocationDesc','Data_Value','Income']]
df_income.head()
```

```python
# creating dummy variable for the income column

income_list = df_income['Income'].unique()

for i in income_list:
    df_income[i] = df_income['Income'].apply(lambda x: int(x==i))
    
# removing the 'data not reported income'

df_income.drop('Data not reported', axis = 1, inplace = True)

df_income
```

```python
# Now let's check the effect of income on obesity

income_cols = df_income.columns[4:9]

X = df_income[income_cols].values
y = df_income['Data_Value'].values

# Now let's split the dataset into train_set and test_set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42, test_size = 0.18)


from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
regressor.coef_
```

```python

plt.barh([0,1,2,3,4], regressor.coef_)
plt.yticks(range(5),df_income.columns[4:9], fontsize = 10)
plt.title('Regression Coefficients')

plt.show()
```

```python
print('R^2 on training...', regressor.score(X_train, y_train))
print('R^2 on test...', regressor.score(X_test, y_test))

print('Model Coefficients', regressor.coef_)
print('Model Intercept', regressor.intercept_)
```
